```python
import string
import os
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = "C:/Users/ABHILASH/Desktop/Information_data"
token_dict = {}
all_stemmed_words = []
ps = PorterStemmer()
stop_words = set(stopwords.words("english")) | set(string.punctuation)
```

```python
for subdir, dirs, files in os.walk(data):
    for file in files:
        path = subdir + os.path.sep + file
        file_contents= open(path,'r')
        if '.txt' in path:
            text = file_contents.read()
            lowercase = text.lower()
            token_dict[file]= lowercase
            file_contents.close()
num_docs = len(token_dict)
doc_names = []
for file_name in token_dict.keys():
    doc_names.append(file_name)
```

```python
# Function to tokenize text
def tokenize_sentence(text):
    return sent_tokenize(text)
def tokenize_word(text):
    return word_tokenize(text)

# Function to remove stop words
def remove_stop_words(words):
    return [w for w in words if w not in stop_words]

# Function for stemming
def stem_words(words):
    return [ps.stem(w) for w in words if w not in stop_words]

# Process each file
for file in token_dict.keys():
    #words = word_tokenize(token_dict[file])
    # Tokenize the content
    Sent_tokens = tokenize_sentence(token_dict[file])
    tokens=tokenize_word(token_dict[file])
    
    print(f"Sentence Tokens for {file}:")
    print(Sent_tokens)
    print()
    print(f"Word Tokens for {file}:")
    print(tokens)
    print()
    
    # Remove stop words
    no_stop_words = remove_stop_words(tokens)
    
    print(f" Stop words removed from {file}:")
    print(no_stop_words)
    print()
    
    # Stem the words
    stemmed_words = stem_words(no_stop_words)

    print(f"Stemmed Tokens for {file} (after removing stop words):")
    print(stemmed_words)
    print()
    all_stemmed_words.append(stemmed_words)
```

```python
# tf -idf
#documents = [" ".join(doc) for doc in all_stemmed_words]
tfidf = TfidfVectorizer(input='all_stemmed_words', stop_words='english')
tfs = tfidf.fit_transform(token_dict.values())
doc_matrix = tfs.toarray()
set_vocab = tfidf.get_feature_names()

# Print the TF-IDF information
print( "TF-IDF document-word matrix.")
#print()
print("%-15s%-20s%-20s%-20s" % ("Words", str(doc_names[0]), str(doc_names[1]), str(doc_names[2])))

for i in range(0, len(set_vocab)):
    print("%-15s%-20s%-20s%-20s" % (str(set_vocab[i]), str(doc_matrix[0][i]), str(doc_matrix[1][i]), str(doc_matrix[2][i])))
```

```python
# cosine similarity

print("Cosine similarity")
for i in range(0, num_docs):
    for j in range(i, num_docs):
        if i != j:
            print("Cosine similarity of %s to %s: %f" % (
                doc_names[i], doc_names[j], cosine_similarity(tfs[i], tfs[j])))
```
